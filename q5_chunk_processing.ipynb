```python
import pandas as pd
import numpy as np

# Step 1: Create LARGE dataset
data = {
    "Region": np.random.choice(["North", "South", "East", "West"], 1000),
    "SalesAmount": np.random.randint(100, 5000, 1000)
}

df = pd.DataFrame(data)

# Save dataset (INPUT FILE)
df.to_csv("large_sales_data.csv", index=False)

# Step 2: Process in CHUNKS
chunk_size = 200
region_totals = {}

for chunk in pd.read_csv("large_sales_data.csv", chunksize=chunk_size):
    grouped = chunk.groupby("Region")["SalesAmount"].sum()

    for region, value in grouped.items():
        if region in region_totals:
            region_totals[region] += value
        else:
            region_totals[region] = value

# Step 3: Convert result to DataFrame
result_df = pd.DataFrame(list(region_totals.items()), columns=["Region", "Total_Sales"])

# Step 4: Save OUTPUT file
result_df.to_csv("q5_output.csv", index=False)

# Step 5: Display result
print(result_df)
```
