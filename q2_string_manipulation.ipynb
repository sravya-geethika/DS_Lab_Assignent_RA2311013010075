```python
import pandas as pd

data = {
    "Employee_Name": [
        "Dr. Rahul Kumar Sharma",
        "Priya M.",
        "Mr. S. Venkatesh",
        "Ms. Anjali Reddy",
        "Rahul Sharma",
        "Dr. Rahul Sharma",
        "Priya M.",
        "Mr. S. Venkatesh"
    ]
}

df = pd.DataFrame(data)
df.head()
```

```python
import re

# Remove titles
df["Clean_Name"] = df["Employee_Name"].str.replace(r"Dr\.|Mr\.|Ms\.", "", regex=True).str.strip()

# Split names
def split_name(name):
    parts = name.split()

    first = parts[0] if len(parts) > 0 else ""
    middle = parts[1][0] if len(parts) > 2 else ""
    last = parts[-1] if len(parts) > 1 else ""

    return pd.Series([first, middle, last])

df[["First_Name", "Middle_Initial", "Last_Name"]] = df["Clean_Name"].apply(split_name)

# Create username
df["Username"] = (df["First_Name"].str[0] + df["Last_Name"]).str.lower()

# Handle duplicates
counts = {}
duplicate_count = 0

def make_unique(username):
    global duplicate_count
    if username in counts:
        counts[username] += 1
        duplicate_count += 1
        return f"{username}{counts[username]}"
    else:
        counts[username] = 1
        return username

df["Username"] = df["Username"].apply(make_unique)

print("Duplicate usernames handled:", duplicate_count)

df.head(8)
```
